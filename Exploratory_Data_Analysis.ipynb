```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
import warnings
sns.set()
```

```python
#Load iris data set file
iris_data=pd.read_csv('iris_data.csv')
iris_data
```

```python
# Display basic information
print("Shape of the dataset:", iris_data.shape)
print("\nData types and missing values:")
print(iris_data.info())
print("\nFirst five rows of the dataset:")
print(iris_data.head(8))
```

```python
#Statistical Insight
iris_data.describe()
```

```python
#Checking For Duplicate Entries
iris_data[iris_data.duplicated()]
```

```python
#Checking the balance
iris_data['species'].value_counts()
```

```python
#Data Visualization
```

```python
#Histogram
iris_data.hist(edgecolor='black',linewidth=1.2,figsize=(12,8))
plt.suptitle('Histogram of Numerical Attributes',size=16)
plt.show()
```

```python
# Suppress specific warning
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn._oldcore")
#Pair Plot
sns.pairplot(iris_data,hue='species')
plt.suptitle('Pair Plot of Iris Dataset\n', size=16)
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(12,8))
for i, column in enumerate(iris_data.columns[:-1],1):
    plt.subplot(2,2,i)
    sns.boxplot(data=iris_data,x='species',y=column)
    plt.title(f'Box Plot of {column}')

plt.tight_layout()
plt.show()
```

```python
#Violin Plots
plt.figure(figsize=(12,8))
for i, column in enumerate(iris_data.columns[:-1],1):
    plt.subplot(2,2,i)
    sns.violinplot(data=iris_data, x='species', y=column)
    plt.title(f'Violin Plot of {column}')

plt.tight_layout()
plt.show()
```

```python

# Compute the correlation matrix, excluding the 'species' column
data = iris_data.drop(['species'], axis=1).corr()

# Display the correlation matrix
print(data)

# Heatmap
sns.heatmap(data, annot=True)

plt.show()
```
